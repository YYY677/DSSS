```python
import pandas as pd
import geopandas as gpd
import os
import numpy as np
```

```python
# 获取文件夹中所有的 shapefile 文件
shapefiles = [os.path.join("./data/LB_shp/", file) for file in os.listdir("./data/LB_shp/") if file.endswith('.shp')]

# 读取所有 shapefile 并合并到一个 GeoDataFrame 中
gdf_list = [gpd.read_file(shp) for shp in shapefiles]
merged_gdf = gpd.GeoDataFrame(pd.concat(gdf_list, ignore_index=True))
merged_gdf
```

```python
merged_gdf = merged_gdf[['lsoa21cd','lsoa21nm','geometry']]

# 根据 lsoa21cd 列进行排序，并丢弃旧索引
merged_gdf = merged_gdf.sort_values(by='lsoa21cd').reset_index(drop=True)
merged_gdf
```

```python
# 将合并后的 GeoDataFrame 保存为 GeoJSON 文件
merged_gdf.to_file("LSOA.gpkg", driver='gpkg')
```
